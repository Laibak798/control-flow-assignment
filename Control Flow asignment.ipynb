```python
#17.Write a Python program to calculate the roots of a quadratic equation.
```

```python
import cmath  # Import to handle complex roots

def calculate_roots(a, b, c):
    """Calculate the roots of a quadratic equation ax^2 + bx + c = 0."""
    if a == 0:
        raise ValueError("Coefficient 'a' cannot be 0 for a quadratic equation.")

    discriminant = b**2 - 4*a*c  # Calculate the discriminant
    root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
    root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)
    return root1, root2

try:
    print("Quadratic Equation: ax^2 + bx + c = 0")
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
    c = float(input("Enter the coefficient c: "))

    root1, root2 = calculate_roots(a, b, c)
    print("\nThe roots of the quadratic equation are:")
    print(f"Root 1: {root1}")
    print(f"Root 2: {root2}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def is_prime(number):
    if number <= 1:
        return f"{number} is not a prime number."
    else:
        if number == 2:
            return f"{number} is a prime number."
        else:
            if number % 2 == 0:
                return f"{number} is not a prime number."
            else:
                i = 3
                while i * i <= number:
                    if number % i == 0:
                        return f"{number} is not a prime number."
                    i += 2
                return f"{number} is a prime number."

try:
    user_input = int(input("Enter a number: "))
    print(is_prime(user_input))
except ValueError:
    print("Invalid input! Please enter a valid integer.")
```

```python
#9. Create a program to calculate the total cost of a shopping cart based on discounts.
```

```python
def calculate_total(cart, discounts):
    """
    Calculate the total cost of items in a shopping cart with discounts applied.

    Parameters:
    cart (list of tuples): Each tuple contains the item name, price, and quantity.
    discounts (dict): A dictionary where keys are item names and values are discounts (in percentage).

    Returns:
    float: The total cost after discounts.
    """
    total_cost = 0

    for item, price, quantity in cart:
        discount = discounts.get(item, 0)  # Get discount for the item, default to 0
        discount_amount = (price * discount / 100) * quantity
        total_cost += (price * quantity) - discount_amount

    return total_cost

shopping_cart = [
    ("apple", 2.00, 5),   # 5 apples at $2.00 each
    ("banana", 1.50, 6),  # 6 bananas at $1.50 each
    ("milk", 3.00, 2),    # 2 bottles of milk at $3.00 each
]

discounts = {
    "apple": 10,  # 10% discount on apples
    "milk": 5,    # 5% discount on milk
}

total = calculate_total(shopping_cart, discounts)
print(f"The total cost of the shopping cart is: ${total:.2f}")
```

```python
def calculate_average(numbers):
  if len(numbers) <= 2:
    return "The list should have at least 3 numbers"

  sorted_numbers = sorted(numbers)
  trimmed_numbers = sorted_numbers[1:-1]
  average = sum(trimmed_numbers)/len(trimmed_numbers)
  return average

numbers = [10,3,5,6,8,9]
result = calculate_average(numbers)
print(f"The average excluding the smallest and largest values is: {result}")
```

```python
def get_matrix(rows, cols):
    matrix = []
    print(f"Enter the elements of a {rows}x{cols} matrix:")
    for i in range(rows):
        row = list(map(int, input(f"Row {i+1}: ").split()))
        if len(row) != cols:
            print(f"Please enter exactly {cols} integers for this row.")
            return get_matrix(rows, cols)
        matrix.append(row)
    return matrix

def print_matrix(matrix):
    """Function to print the matrix."""
    for row in matrix:
        print(' '.join(map(str, row)))

def add_matrices(matrix1, matrix2):
    """Function to add two matrices."""
    return [[matrix1[i][j] + matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]

def subtract_matrices(matrix1, matrix2):
    """Function to subtract the second matrix from the first matrix."""
    return [[matrix1[i][j] - matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]

def main():
    rows = int(input("Enter the number of rows for the matrices: "))
    cols = int(input("Enter the number of columns for the matrices: "))

    print("\nMatrix 1:")
    matrix1 = get_matrix(rows, cols)

    print("\nMatrix 2:")
    matrix2 = get_matrix(rows, cols)

    print("\nChoose an operation:")
    print("1. Add Matrices")
    print("2. Subtract Matrices")
    choice = input("Enter your choice (1/2): ")

    if choice == '1':
        result = add_matrices(matrix1, matrix2)
        print("\nResultant Matrix (Addition):")
        print_matrix(result)
    elif choice == '2':
        result = subtract_matrices(matrix1, matrix2)
        print("\nResultant Matrix (Subtraction):")
        print_matrix(result)
    else:
        print("Invalid choice! Please choose 1 or 2.")

if __name__ == "__main__":
    main()
```

```python
def matrix_multiply(A, B):
    if len(A[0]) != len(B):
        return "Matrices cannot be multiplied due to dimension mismatch."

    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    for i in range(len(A)):  # Iterate over rows of A
        for j in range(len(B[0])):  # Iterate over columns of B
            for k in range(len(B)):  # Iterate over rows of B
                result[i][j] += A[i][k] * B[k][j]

    return result

# Example usage:
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

result = matrix_multiply(A, B)
print("Resultant Matrix:")
for row in result:
    print(row)
```

```python
#35. Create a program that simulates a basic text-based tic-tac-toe game against the computer
```

```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 5)

def check_win(board, mark):
    for i in range(3):
        if all([cell == mark for cell in board[i]]) or all([board[j][i] == mark for j in range(3)]):
            return True
    if all([board[i][i] == mark for i in range(3)]) or all([board[i][2-i] == mark for i in range(3)]):
        return True
    return False

def check_draw(board):
    return all([cell != ' ' for row in board for cell in row])

def player_move(board):
    while True:
        move = input("Enter your move (1-9): ")
        if move.isdigit() and 1 <= int(move) <= 9:
            row, col = divmod(int(move) - 1, 3)
            if board[row][col] == ' ':
                board[row][col] = 'X'
                break
            else:
                print("That spot is already taken!")
        else:
            print("Invalid input. Please enter a number between 1 and 9.")

def computer_move(board):
    available_moves = [(i, j) for i in range(3) for j in range(3) if board[i][j] == ' ']
    row, col = random.choice(available_moves)
    board[row][col] = 'O'

def tic_tac_toe():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic-Tac-Toe! You are 'X' and the computer is 'O'.")

    while True:
        print_board(board)
        player_move(board)
        if check_win(board, 'X'):
            print_board(board)
            print("Congratulations, you win!")
            break
        if check_draw(board):
            print_board(board)
            print("It's a draw!")
            break

        computer_move(board)
        if check_win(board, 'O'):
            print_board(board)
            print("Computer wins! Better luck next time.")
            break
        if check_draw(board):
            print_board(board)
            print("It's a draw!")
            break

tic_tac_toe()
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False

def get_days_in_month(month, year):
    if month in {1, 3, 5, 7, 8, 10, 12}:  # Months with 31 days
        return 31
    elif month in {4, 6, 9, 11}:  # Months with 30 days
        return 30
    elif month == 2:  # February
        return 29 if is_leap_year(year) else 28
    else:
        return 0  # Invalid month

def get_start_day(month, year):
    if month < 3:
        month += 12
        year -= 1
    k = year % 100
    j = year // 100
    day = (1 + 13 * (month + 1) // 5 + k + k // 4 + j // 4 - 2 * j) % 7
    return (day + 6) % 7  # Adjust to make 0=Sunday, 1=Monday, ..., 6=Saturday

def print_calendar(month, year):
    days_in_month = get_days_in_month(month, year)
    start_day = get_start_day(month, year)

    print(f"\n    Calendar for {month}/{year}")
    print("Su Mo Tu We Th Fr Sa")

    print("   " * start_day, end="")

    for day in range(1, days_in_month + 1):
        print(f"{day:2}", end=" ")
        if (start_day + day) % 7 == 0:  # New line after Saturday
            print()
    print()

month = int(input("Enter month (1-12): "))
year = int(input("Enter year: "))

if 1 <= month <= 12:
    print_calendar(month, year)
else:
    print("Invalid month. Please enter a value between 1 and 12.")
```

```python
#39. Build a program that simulates a basic text-based blackjack game against the computer.
```

```python
import random

def deal_card():
    """Simulates dealing a card. Cards are between 1 and 11."""
    return random.randint(1, 11)

def calculate_score(hand):
    """Calculates the score of a hand."""
    score = sum(hand)
    # Adjust for Ace (11 becomes 1 if the score exceeds 21)
    if 11 in hand and score > 21:
        hand[hand.index(11)] = 1
        score = sum(hand)
    return score

def display_hand(hand, owner):
    """Displays the current hand of the player or computer."""
    print(f"{owner} hand: {hand} | Total: {sum(hand)}")

def blackjack():
    print("Welcome to Text-Based Blackjack!")

    player_hand = [deal_card(), deal_card()]
    computer_hand = [deal_card(), deal_card()]

    while True:
        display_hand(player_hand, "Your")
        player_score = calculate_score(player_hand)
        if player_score > 21:
            print("You went over 21! You lose.")
            return
        choice = input("Do you want to (H)it or (S)tand? ").lower()
        if choice == 'h':
            player_hand.append(deal_card())
        elif choice == 's':
            break
        else:
            print("Invalid choice. Please type 'H' or 'S'.")

    print("\nComputer's turn...")
    while calculate_score(computer_hand) < 17:
        computer_hand.append(deal_card())

    display_hand(player_hand, "Your")
    display_hand(computer_hand, "Computer")

    player_score = calculate_score(player_hand)
    computer_score = calculate_score(computer_hand)

    if computer_score > 21 or player_score > computer_score:
        print("You win!")
    elif player_score == computer_score:
        print("It's a draw!")
    else:
        print("You lose!")

blackjack()
```

```python
#40.Write a program that generates the prime factors of a given number using trial division.
```

```python
def prime_factors(n):
    """Generate the prime factors of a given number using trial division."""
    factors = []
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    factor = 3
    while factor * factor <= n:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
        factor += 2

    if n > 1:
        factors.append(n)

    return factors

number = int(input("Enter a number to find its prime factors: "))
if number <= 0:
    print("Please enter a positive integer greater than 0.")
else:
    print("Prime factors of", number, "are:", prime_factors(number))
```
